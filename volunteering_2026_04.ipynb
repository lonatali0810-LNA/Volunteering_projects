**Определяем регионы с наибольшим количеством зарегистрированных доноров**

```python
# Определяем регионы с наибольшим количеством зарегистрированных доноров
SELECT 
    region, 
    COUNT(id) AS user_count
FROM donorsearch.user_anon_data
GROUP BY region
ORDER BY user_count DESC;
```

Вывод: на первом месте по зарегистрированным донорам лидируют самые крупные города России: Москва, Санкт-петербург, Казань, Киев (Украина) и др. и много пропущенных значений NULL - 100574

```python
# Определим в каких регионах активнее сдают кровь, выявим регионы с наибольшим донорским потенциалом
SELECT
    region,
    SUM(CASE WHEN blood_class = 'Цельная кровь' THEN 1 ELSE 0 END) AS whole_blood_donations,
    SUM(CASE WHEN blood_class = 'Плазма' THEN 1 ELSE 0 END) AS plasma_donations,
    SUM(CASE WHEN blood_class = 'Тромбоциты' THEN 1 ELSE 0 END) AS platelets_donations,
    SUM(CASE WHEN blood_class = 'Эритроциты' THEN 1 ELSE 0 END) AS red_cells_donations,
    SUM(CASE WHEN blood_class = 'Гранулоциты (Лейкоциты)' THEN 1 ELSE 0 END) AS white_cells_donations,
    SUM(CASE WHEN blood_class = 'Костный мозг' THEN 1 ELSE 0 END) AS bone_marrow_donations,
    COUNT(*) AS total_donations
FROM donorsearch.donation_anon
GROUP BY region
ORDER BY total_donations DESC;
```

Вывод: при этом активнее всего сдают кровь Татарстан, Ханты-Мансийский автономный округ, Башкортостан, Свердловская область

```python
# Посмотрим какую кровь сдают доноры и в каком количестве в разрезе регионов
SELECT 
    region, 
    blood_class, 
    COUNT(*) AS count_total
FROM donorsearch.donation_anon
GROUP BY region, blood_class;
```

Вывод: чаще всего люди сдают цельную кровь, потом сдают плазму, тромбоциты. 

**Определим влияние пола и возраста на донации**

```python
# Выявим влияние возраста на количество донаций
# Расмотрим количество мужчин и женщин в системе о пользователях
SELECT 
    COUNT(id), 
    gender 
FROM donorsearch.user_anon_data
GROUP BY gender;
```

Вывод: женщины чаще регистрируются в качестве доноров крови (54933), чем мужчины (48952)

```python
# Определим какой пол чаще сдает кровь
SELECT 
    COUNT(dn.id) AS total_donation, 
    an.gender
FROM donorsearch.donation_anon AS dn 
JOIN donorsearch.user_anon_data AS an ON dn.user_id=an.id 
GROUP BY an.gender;
```

Вывод: однако, мужчины чаще сдают кровь (123920), чем женщины (72224).

```python
# Определим как влияет возраст на донации
WITH valid_users AS (
    # Сначала отбираем корректные записи с датой рождения
    SELECT
        id,
        birth_date,
        EXTRACT(YEAR FROM CURRENT_DATE) - EXTRACT(YEAR FROM birth_date) AS age
    FROM donorsearch.user_anon_data
    WHERE birth_date IS NOT NULL  -- Исключаем NULL
      AND EXTRACT(YEAR FROM birth_date) <= EXTRACT(YEAR FROM CURRENT_DATE)  -- Проверяем, что дата рождения не в будущем
      AND birth_date >= '1900-01-01'  -- Дополнительно фильтруем явно некорректные даты
),
age_groups AS (
    # Группируем пользователей по возрастным категориям
    SELECT
        id,
        CASE
            WHEN age BETWEEN 18 AND 25 THEN '18–25'
            WHEN age BETWEEN 26 AND 35 THEN '26–35'
            WHEN age BETWEEN 36 AND 45 THEN '36–45'
            WHEN age BETWEEN 46 AND 55 THEN '46–55'
            ELSE '56+'
        END AS age_group
    FROM valid_users
)
# Основной запрос: считаем донации по возрастным группам
SELECT
    ag.age_group,
    COUNT(DISTINCT ag.id) AS donor_count,  -- Количество уникальных доноров в группе
    COUNT(da.id) AS total_donations,       -- Общее количество донаций в группе
    ROUND(COUNT(da.id) * 1.0 / COUNT(DISTINCT ag.id), 2) AS avg_donations  -- Среднее количество донаций на донора
FROM age_groups ag
LEFT JOIN donorsearch.donation_anon da ON ag.id = da.user_id
GROUP BY ag.age_group
ORDER BY ag.age_group;
```

Вывод: возраст влияет на донации. Категория 36-45 лет не самая многочисленая, но зато самая мотивированная (среднее кол-во донаций на человека 2,68). Именно они больше всего сдают донации. 
При этом самой многочисленной группой являются 26-35 лет, но среднее кол-во донаций на донора составляет лишь 1,35%. 

Рекомендации: самые мотивированные явлются в категории 36-45 и 46-55 лет. Молодые люди в возрасте 26-35 лет активно записываются, но не доходят до сдачи донаций.Необходимо поработать именно с этой категорией молодых людей.

**Динамика общего количества донаций в месяц за 2022 и 2023 годы**

```python
# Посмотрим на динамику общего количества донаций в месяц за 2022 и 2023 годы
SELECT 
    DATE_TRUNC('month', donation_date) AS month, 
    COUNT(id) AS id_count
FROM donorsearch.donation_anon
WHERE donation_date BETWEEN '2022-01-01' AND '2023-12-31'
GROUP BY MONTH
ORDER BY month;
```

**Влияние системы бонусов влияет на зарегистрированные в системе донации**

```python
# Оценить, как система бонусов влияет на зарегистрированные в системе донации
WITH donor_activity AS
  (SELECT u.id,
          u.confirmed_donations,
          COALESCE(b.user_bonus_count, 0) AS user_bonus_count
   FROM donorsearch.user_anon_data u
   LEFT JOIN donorsearch.user_anon_bonus b ON u.id = b.user_id)
SELECT CASE
           WHEN user_bonus_count > 0 THEN 'Пользователи с бонусами'
           ELSE 'Пользователи без бонусов'
       END AS статус_бонусов,
       COUNT(id) AS количество_доноров,
       AVG(confirmed_donations) AS среднее_количество_донаций
FROM donor_activity
GROUP BY статус_бонусов;
```

Вывод: большинство зарегистрированных пользователей (256491) без бонусов, при этом среднее количество подтвержденных донаций всего 0,53. А зарегситрированные пользователи с бонусами - 21108, но их средний уровень донаций составляет 13,90. Что подтверждает положительное влияние программ лояльности на активность доноров.

**Определить каналы привлечения донорства (сколько по каким каналам пришло доноров, среднее количество донаций по каждому каналу)**

```python
SELECT CASE WHEN autho_vk THEN 'VKontakte'
            WHEN autho_ok THEN 'Odnoklassniki'
            WHEN autho_tg THEN 'Telegram'
            WHEN autho_yandex THEN 'Yandex'
            WHEN autho_google THEN 'Google'
            ELSE 'Другие способы авторизации'
          END AS social_channel,
          COUNT(id) AS count_id,
          ROUND(AVG(confirmed_donations), 2) AS avd_confirmed_donations
FROM donorsearch.user_anon_data
GROUP BY social_channel
ORDER BY count_id DESC;
```

**Сравнить активность однократных доноров со средней активностью повторных доноров**

```python
WITH donor_groups AS ( # определим однократных и повторынх доноров
    SELECT
        id AS donor_id,
        confirmed_donations AS con_donations,
        CASE
            WHEN confirmed_donations = 1 THEN 'Однократные доноры'
            ELSE 'Повторные доноры'
        END AS donor_group
    FROM donorsearch.user_anon_data
    WHERE confirmed_donations > 0)

SELECT
    donor_group,
    COUNT(donor_id) AS count_of_donors,
    SUM(con_donations) AS sum_donations,
    ROUND(AVG(con_donations), 2) AS average_donations_per_donor
FROM donor_groups
GROUP BY donor_group
ORDER BY donor_group;
```

ВЫВОД: видим, что количество активных доноров делит примерно на равную группу - однократных и многократных доноров (19510 и 19127 соответственно). Но при этом отмечаем, что активность повторных доноров в разы выше однократных доноров - 19510 и 203367 (по сумме) и 1 и 10,63 (по средней активности). Таким образом, повторные доноры это ключевая аудитория, с которыми нужно работать и которых необходимо поддерживать.

**Сравнить данные о планируемых донациях с фактическими данными, чтобы оценить эффективность планирования**

```python
WITH planned_donations AS ( # планируемые донации
    SELECT DISTINCT user_id, donation_date, donation_type
    FROM donorsearch.donation_plan),
  
actual_donations AS ( # фактические донации
    SELECT DISTINCT user_id, donation_date
    FROM donorsearch.donation_anon),
  
planned_vs_actual AS ( # объединение двух вспомогательных таблиц
  SELECT
    pd.user_id,
    pd.donation_date AS planned_date,
    pd.donation_type,
    CASE WHEN ad.user_id IS NOT NULL THEN 1 ELSE 0 END AS completed
  FROM planned_donations pd
  LEFT JOIN actual_donations ad ON pd.user_id = ad.user_id AND pd.donation_date = ad.donation_date)
  
SELECT # расчет количества запланированных, фактический донаций
  donation_type,
  COUNT(*) AS total_planned_donations,
  SUM(completed) AS completed_donations,
  ROUND(SUM(completed) * 100.0 / COUNT(*), 2) AS completion_rate
FROM planned_vs_actual
GROUP BY donation_type;
```

Вывод: видим, что количество запланированных "безвозмездно" в разы больше, чем платно (22 903 против 3299). Однако, количество выполненных донаций очень снижается (4950 и 429 соответственно). Отсюда и процент выполнения планов донаций низок для обоих типов доноров: 21.61% для безвозмездных и 13.00% для платных.
